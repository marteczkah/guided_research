```python
import numpy as np  
import matplotlib.pyplot as plt  
import h5py
import plotly.express as px
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import transforms
from torch.utils.data import Dataset, DataLoader
```

```python
def plot_cloud_no_label(points):
  my_list = []
  for i, point in enumerate(points):
    x = point[0]
    y = point[1]
    z = point[2]
    cur_dic = {'x': x, 'y': y, 'z': z}
    my_list.append(cur_dic)
  import pandas as pd
  data_frame = pd.DataFrame(my_list)  
  import plotly.express as px
  fig = px.scatter_3d(data_frame, x='x', y='y', z='z')
  fig.show()
```

# Critical Points -> Classification

```python
seg_features = h5py.File('/content/gdrive/MyDrive/GR_Data/hx_clas.h5')
```

```python
seg_features.keys()
```

```python
seg_features['pts'][0]
```

```python
hxs = seg_features['hx']
```

```python
g_f_points = []
for hx in hxs:
  indx = np.argmax(hx, axis=1)
  g_f_points.append(indx)
```

```python
cp = []
for i in range(len(g_f_points)):
  cp.append(seg_features['pts'][i][g_f_points[i]])
```

```python
plot_cloud_no_label(cp[3])
```

```python
new_file = h5py.File("chair_cp.h5", "w")
```

```python
centroids_dset = new_file.create_dataset("pts", data=seg_features['pts'], compression='gzip', compression_opts=4, )
centroids_dset = new_file.create_dataset("cp", data=cp, compression='gzip', compression_opts=4, )
centroids_dset = new_file.create_dataset("gf", data=seg_features['gf'], compression='gzip', compression_opts=4, )
```
